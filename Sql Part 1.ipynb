```python
import pandas as pd
import sqlite3 as sql
```

```python
conn = sql.connect('scopus_data.sqlite')
```

### 1.	Tampilkan semua kolom dari AI_scopus yang terbit pada tahun 2016

```python
df1 = pd.read_sql_query("SELECT * FROM AI_scopus WHERE Year = '2016'", conn)
df1
```

### 2.	Tampilkan semua kolom dari AI_scopus yang terbit pada tahun 2009 hingga 2018

```python
df2 = pd.read_sql_query("SELECT * FROM AI_scopus WHERE Year>=2009 AND Year<=2018", conn)
```

```python
df2
```

### 3.	Tampilkan kolom LanguageofOriginalDocument dari AI_scopus

```python
df3 = pd.read_sql_query("SELECT LanguageofOriginalDocument FROM AI_scopus", conn)
df3
```

### 4.	Tentukan berapa banyak bahasa pada kolom LanguageofOriginalDocument

```python
df4 = pd.read_sql_query("SELECT LanguageofOriginalDocument, COUNT(DISTINCT LanguageofOriginalDocument) FROM AI_scopus GROUP BY LanguageofOriginalDocument", conn)
df4
```

### 5.	Tampilkan judul dari semua artikel yang terbit pada “AAAI Conf. Artif. Intell., AAAI”.

```python
df5 = pd.read_sql_query("SELECT Title FROM AI_scopus WHERE AbbreviatedSourceTitle = 'AAAI Conf. Artif. Intell., AAAI'", conn)
df5
```

### 6.	Tentukan berapa banyak publisher yang mempublikasikan artikel pada tahun 2009 hingga 2012

```python
df6 = pd.read_sql_query("SELECT Year,COUNT(Publisher) FROM AI_scopus WHERE Year>=2009 AND Year<=2012 GROUP BY Year",conn)
```

```python
df6
```

### 7.	Artikel mana yang paling tebal dari AI_scopus?

```python
df7 = pd.read_sql_query("SELECT Title,MAX(Pagecount) FROM AI_scopus", conn)
```

```python
df7
```

### 8.	Berapa banyak artikel yang dipublish oleh penulis dari IIT-Bombay dari tahun 2009 hingga 2013?

```python
df8 = pd.read_sql_query("SELECT Year, COUNT(Title) FROM AI_scopus WHERE (Year>=2009 AND Year<=2013) AND (Affiliations LIKE '%Bombay%') GROUP BY Year",conn)
df8
```

### 9.	Berapa banyak judul yang masuk ke dalam golongan “Article” dan berapa banyak yang masuk ke dalam golongan “Conference Paper”?

```python
df9 = pd.read_sql_query("SELECT DocumentType, COUNT(DocumentType) FROM AI_scopus WHERE DocumentType='Article' OR DocumentType='Conference Paper' GROUP BY DocumentType", conn)
df9
```

### 10.	Tentukan paper dengan cited_rank tertinggi!

```python
df10 = pd.read_sql_query("SELECT Title, MAX(cited_rank) FROM AI_scopus", conn)
df10
```

### 11.	Berapa banyak paper yang memiliki cited_rank diatas 70

```python
df11 = pd.read_sql_query("SELECT COUNT(cited_rank) FROM AI_scopus WHERE cited_rank>70", conn)
df11
```
